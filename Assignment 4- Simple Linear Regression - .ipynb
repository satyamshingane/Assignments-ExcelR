```python
#Simple linear Regression
df1 = df.rename({'Delivery Time':'Delivery_Time','Sorting Time':'Sorting_Time'}, axis = 1)
df1
model = sms.ols('Delivery_Time~Sorting_Time', data = df1).fit()
```

```python
model.summary()
```

Conclusion-
The R-squared and Adjusted R-squared scores are lower than 0.9
Hence, transformation is needed.

# Model for Log transformation

```python
model_new = sms.ols('np.log(Delivery_Time)~np.log(Sorting_Time)', data = df1).fit()
model_new.summary()
```

Conclusion-
The R-squared and Adjusted R-squared scores are lower than 0.9

# model for Square Root transformation

```python
model1 = sms.ols('np.sqrt(Delivery_Time)~np.sqrt(Sorting_Time)', data = df1).fit()
model1.summary()
```

Conclusion-
The R-squared and Adjusted R-squared scores are increased but still lower than 0.9

# Model for Reciprocal transformation

```python
D1=1/(df['Delivery Time'])
S1=1/(df['Sorting Time'])
model2 = sms.ols('D1~S1', data = df1).fit()
model2.summary()
```

Conclusion-
The R-squared and Adjusted R-squared scores are increased but still lower than 0.9

# Model for square transformation

```python
D2=(df['Delivery Time'])*(df['Delivery Time'])
S2=(df['Sorting Time'])*(df['Sorting Time'])
model3 = sms.ols('D2~S2', data = df1).fit()
model3.summary()
```

Conclusion-
The R-squared and Adjusted R-squared scores are still lower than 0.9

# After performing 4 transformation on the Data, the Reciprocal Transformation Model has scored the highest R-squared and Adjusted R-squared scores than the other models.
Adj. R-squared =0.827 
R-squared =0.835
